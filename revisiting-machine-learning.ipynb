```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings ('ignore')

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv(r'learningSet.txt')
df
```

```python
df.isna().sum()
```

```python
categorical_columns = df.select_dtypes(exclude=['number']).columns.tolist()
categorical_columns
```

```python
drop_list = []
drop_list.append('OSOURCE')
drop_list.append('ZIP')
df.drop(columns = drop_list, inplace = True)
```

```python
null_df = df.isna().sum()/len(df)
null_df
```

```python
null_df = pd.DataFrame(null_df).reset_index()
null_df.columns = ['column_names', 'percentage_of_nulls']
null_df
```

```python
null_df = null_df.sort_values(by='percentage_of_nulls', ascending=False )
null_df
```

```python
threshold = 0.85

condition = null_df['percentage_of_nulls'] > threshold
columns_above_threshold = null_df[condition]
columns_above_threshold
```

```python
drop_column_list = list(columns_above_threshold['column_names'])
df = df.drop(columns = drop_column_list)
```

```python
df['GENDER'] = df['GENDER'].fillna('F')
df['GENDER'] = df['GENDER'].apply(lambda x: x if x in ['M', 'F'] else 'other')
```
